```python
#importing pandas for data preprocessing
import pandas as pd
#mlxtend for association algorithms
from mlxtend.frequent_patterns import apriori,association_rules
from mlxtend.preprocessing import TransactionEncoder
#matplotlib for visualisations
import matplotlib.pyplot as plt
```

```python
book = pd.read_csv("book.csv")
book.head()
```

```python
#Setting min_support and min_threshold to different values
frequent_itemsets1 = apriori(book, min_support=0.1, use_colnames=True)
frequent_itemsets2 = apriori(book, min_support=0.2, use_colnames=True)
frequent_itemsets3 = apriori(book, min_support=0.1, use_colnames=True)
```

```python
rules1 = association_rules(frequent_itemsets1, metric="lift", min_threshold=0.7)
rules2 = association_rules(frequent_itemsets2, metric="lift", min_threshold=0.7)
rules3 = association_rules(frequent_itemsets3, metric="lift", min_threshold=0.9)
```

```python
rules1.sort_values('lift',ascending = False)[0:20]
```

```python
rules2.sort_values('lift',ascending = False)[0:20]
```

```python
rules3.sort_values('lift',ascending = False)[0:20]
```

```python
#plots for min_support=0.1 and min_threshold=0.7
plt.scatter(rules1['support'], rules1['confidence'], alpha=0.5)
plt.xlabel('support')
plt.ylabel('confidence')
plt.title('Support vs Confidence')
plt.show()
```

```python
plt.scatter(rules1['support'], rules1['lift'], alpha=0.5)
plt.xlabel('support')
plt.ylabel('lift')
plt.title('Support vs lift')
plt.show()
```
