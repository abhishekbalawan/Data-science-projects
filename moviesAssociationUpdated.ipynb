```python
#importing pandas for data preprocessing
import pandas as pd
#mlxtend for association algorithms
from mlxtend.frequent_patterns import apriori,association_rules
from mlxtend.preprocessing import TransactionEncoder
#matplotlib for visualisations
import matplotlib.pyplot as plt
```

```python
movies = pd.read_csv("my_movies.csv")
movies.columns
```

```python
movies=movies[['Sixth Sense', 'Gladiator', 'LOTR1', 'Harry Potter1', 'Patriot', 'LOTR2', 'Harry Potter2', 'LOTR', 'Braveheart', 'Green Mile']]
movies
```

```python
#Setting min_support and min_threshold to different values
frequent_itemsets1 = apriori(movies, min_support=0.1, use_colnames=True)
frequent_itemsets2 = apriori(movies, min_support=0.2, use_colnames=True)
frequent_itemsets3 = apriori(movies, min_support=0.3, use_colnames=True)
frequent_itemsets4 = apriori(movies, min_support=0.4, use_colnames=True)
frequent_itemsets5 = apriori(movies, min_support=0.4, use_colnames=True)
```

```python
rules1 = association_rules(frequent_itemsets1, metric="lift", min_threshold=0.7)
rules2 = association_rules(frequent_itemsets2, metric="lift", min_threshold=0.7)
rules3 = association_rules(frequent_itemsets3, metric="lift", min_threshold=0.9)
rules4 = association_rules(frequent_itemsets4, metric="lift", min_threshold=0.6)
rules5 = association_rules(frequent_itemsets5, metric="lift", min_threshold=0.5)
```
