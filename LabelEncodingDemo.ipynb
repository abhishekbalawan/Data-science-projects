```python
from mpl_toolkits import mplot3d
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
fig = plt.figure()
ax = plt.axes(projection='3d')
```

```python
import pandas as pd
   
data1 = {'x1': [20,21,23,22,2,3,4,1],
        'x2': [25,20,19,23,1,3,2,4],
        'y' : [1,2,2.2,1.6,12,10,11,9]
       }
  
df1 = pd.DataFrame(data1,columns=['x1','x2','y'])
print (df1)
```

```python
ax = plt.axes(projection='3d')
zdata = df1['y']
xdata = df1['x1']
ydata = df1['x2']
ax.scatter3D(xdata, ydata, zdata, c=zdata);
```
