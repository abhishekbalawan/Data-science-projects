```python
from keras.models import Sequential
from keras.layers import Dense
```

```python
model = Sequential()
model.add(Dense(12, input_dim=8, activation='relu'))
model.add(Dense(8, activation='relu'))
model.add(Dense(1, activation='sigmoid'))
```

```python
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
from google.colab import files
uploaded = files.upload()
```

```python
import io
import pandas as pd
gasTurbines = pd.read_csv(io.BytesIO(uploaded['gas_turbines.csv']))
```

```python
#storing inputs into X; removing columns Turbine Inlet Temperature and Compressor Discharge pressure as they are turbine parameters and not ambient parameters
X=gasTurbines[['AT', 'AP', 'AH', 'AFDP', 'GTEP', 'TIT',  'CO','NOX']]
#Storing output into Y
Y=gasTurbines[['TEY']]
```

```python
X
```

```python
Y
```

```python
model.fit(X, Y, epochs=1000, batch_size=10000)
```

```python
loss, accuracy = model.evaluate(X, Y)
print('Accuracy: %.2f' % (accuracy*100))
print('loss: %d' % loss)
```
